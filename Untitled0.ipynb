```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [300, 600, 750],
    'max_depth':    [10, 12, 15],
    'max_features': [None, 'sqrt', 5],
    'min_samples_leaf': [1],
}

gs = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    scoring='neg_mean_squared_error',
    cv=10,
    n_jobs=-1,
    verbose=1,
)
gs.fit(TRAIN_X, TRAIN_Y)

model = gs.best_estimator_
print(f"\nBest params: {gs.best_params_}")
print(f"Best CV MSE: {-gs.best_score_:.6e}")
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(EVAL_X)

mse       = mean_squared_error(EVAL_Y, predictions)
rmse      = np.sqrt(mse)
r         = np.corrcoef(EVAL_Y, predictions)[0, 1]
error_pct = np.abs(EVAL_Y - predictions) / EVAL_Y * 100

print("=" * 58)
print(f"  {'Circuit':<10} {'Actual':>9} {'Predicted':>11} {'Error%':>8}")
print("  " + "-" * 42)
for name, a, p, e in zip(EVAL_CIRCUITS, EVAL_Y, predictions, error_pct):
    print(f"  {name:<10} {a:>9.5f} {p:>11.5f} {e:>7.2f}%")
print("  " + "-" * 42)
print(f"  MSE  : {mse:.6e}   (paper: 1.46e-06)")
print(f"  RMSE : {rmse:.6f}  (paper: 0.000116)")
print(f"  R    : {r:.5f}    (paper: 0.99938)")
print("=" * 58)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Circuit':      EVAL_CIRCUITS,
    'Actual_mW':    EVAL_Y,
    'Predicted_mW': predictions,
    'Error_%':      error_pct.round(2),
})
df.to_csv('predictions.csv', index=False)

from google.colab import files
files.download('predictions.csv')
print("Done — predictions.csv downloaded.")
```
